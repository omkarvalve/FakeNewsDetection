```python
import pandas as pd
```

```python
#api url
general = 'https://newsapi.org/v2/everything?q=bitcoin&apiKey=ec7e54d77d7a4fe08d6144d9c5309bda'
indian_news = 'https://newsapi.org/v2/top-headlines?country=in&apiKey=ec7e54d77d7a4fe08d6144d9c5309bda'
```

```python
#fetching api
import requests
import json
general_APIresponse = requests.get(general)
indian_news_APIresponse = requests.get(indian_news)
#print(response_API.status_code)
data1 = general_APIresponse.text
data2 = indian_news_APIresponse.text
parse_json1 = json.loads(data1)
parse_json2 = json.loads(data2)
```

```python
#convertjsontocsv
csv1=pd.json_normalize(parse_json1['articles'])
csv2=pd.json_normalize(parse_json2['articles'])
csv1.to_csv('test1.csv', index=False, encoding='utf-8')
csv2.to_csv('test2.csv', index=False, encoding='utf-8')
                
```

```python
#readingcsvfile
general=pd.read_csv('test1.csv',usecols = ['title','description','content'])
indian_news=pd.read_csv('test2.csv',usecols = ['title','description','content'])
general.tail()
```

```python
indian_news.tail()
```

```python
#datapreprocessing
common_csv = []
common_csv.extend(general['title'])
common_csv.extend(general['description'])
common_csv.extend(general['content'])
common_csv.extend(indian_news['title'])
common_csv.extend(indian_news['description'])
common_csv.extend(indian_news['content'])
#iteratorfunction
def iterator (len,arr,str):
  for x in range(len):
    arr.append(str)
  return arr

df=pd.DataFrame({'text':common_csv, 'label' :iterator(len(common_csv),[],'REAL')})
df.to_csv('common.csv', index=False, encoding='utf-8')
```
